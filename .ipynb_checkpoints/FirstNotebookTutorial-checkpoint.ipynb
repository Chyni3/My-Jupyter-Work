# The Weather Dataset
Here. The Weather Dataset is a time-series data set with per-hour information about weather conditions at a particular location. it records Temperature, Dew Point Temperature, Relative Humidity, Wind Speed, Visibility, Pressure and Conditions.

This data is available in CSV file. We are going to analyse this data using Pandas DataFrame.

```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\chiku\Downloads\1. Weather Data.csv")
```

```python
data
```

# How to Analyse DataFrames 
# .head()
It shows the first N rows in the data (by default, N=5)

```python
data.head()
```

# .shape
it shows the total number of rows and columns in a dataframe

```python
data.shape
```

# .index
This attribute provides the index of the dataframe

```python
data.index
```

# .columns

```python
data.columns
```

# .dtypes

```python
data.dtypes
```

# .unique()
In a column it shows all the unique values. It can be applied on a single column only, not on the whole dataframe.

```python
data['Weather'].unique()
```

# .nunique()
It shows the total number of unique values in each column. It can be applied on as single column as well as on whole dataframe.

```python
data.nunique()
```

# .count
It shows the total number of non-null in each column. It can be applied on a single column asa well as on whole dataframe.


```python
data.count()
```

# .value_counts
In a column, it shows all the unique values with their count. It can be applied on single column only.

```python
data['Weather'].value_counts()
```

# .info()
Provides basic information about the dataframe.

# data.info()
Provides basic information about the dataframe

```python
data.info()
```
